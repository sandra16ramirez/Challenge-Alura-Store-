### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Facturacion_tienda1 = tienda["Precio"].sum()
Facturacion_tienda2 = tienda2["Precio"].sum()
Facturacion_tienda3 = tienda3["Precio"].sum()
Facturacion_tienda4 = tienda4["Precio"].sum()

print('Facturacion total por tienda \n')
print(f"Tienda 1: ${Facturacion_tienda1:,.2f}")
print(f"Tienda 2: ${Facturacion_tienda2:,.2f}")
print(f"Tienda 3: ${Facturacion_tienda3:,.2f}")
print(f"Tienda 4:e ${Facturacion_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# Lista de facturación por tienda
facturaciones = [Facturacion_tienda1, Facturacion_tienda2, Facturacion_tienda3,Facturacion_tienda4]

# Nombres de las tiendas en el eje X
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(5, 4))
barras = plt.bar(tiendas, facturaciones, color='skyblue')

# Agregar etiquetas de datos encima de cada barra
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 50,  # +50 para que no quede pegado
             f"${altura:,.0f}", ha='center', va='bottom', fontsize=8)

# Títulos y etiquetas
plt.title("Facturación total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación ($)")
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.tight_layout()
```

# 2. Ventas por categoría

```python
Ventas_por_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Categoría del Producto"].value_counts()
Ventas_por_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].value_counts()
Ventas_por_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].value_counts()
Ventas_por_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].value_counts()

print('VENTAS POR CATEGORIA \n')
print("Ventas tienda 1:")
print(Ventas_por_categoria_tienda1.sort_values(ascending=True))

print("\n Ventas tienda 2:")
print(Ventas_por_categoria_tienda2.sort_values(ascending=True))

print("\n Ventas tienda 3:")
print(Ventas_por_categoria_tienda3.sort_values(ascending=True))

print("\n Ventas tienda 4:")
print(Ventas_por_categoria_tienda4.sort_values(ascending=True))

```

```python
Ventas_por_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Precio"].sum()
Ventas_por_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Precio"].sum()
Ventas_por_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Precio"].sum()
Ventas_por_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Precio"].sum()

print('VENTAS POR CATEGORIA \n')
print("Ventas tienda 1:")
print(Ventas_por_categoria_tienda1.sort_values(ascending=True).apply(lambda x: f"${x:,.2f}"))

print("\n Ventas tienda 2:")
print(Ventas_por_categoria_tienda2.sort_values(ascending=True).apply(lambda x: f"${x:,.2f}"))

print("\n Ventas tienda 3:")
print(Ventas_por_categoria_tienda3.sort_values(ascending=True).apply(lambda x: f"${x:,.2f}"))

print("\n Ventas tienda 4:")
print(Ventas_por_categoria_tienda4.sort_values(ascending=True).apply(lambda x: f"${x:,.2f}"))

```

```python
# DataFrame de facturación por categoría (usa las Series ya definidas arriba)
df_facturacion_plot = pd.concat([
    Ventas_por_categoria_tienda1.rename("Tienda 1"),
    Ventas_por_categoria_tienda2.rename("Tienda 2"),
    Ventas_por_categoria_tienda3.rename("Tienda 3"),
    Ventas_por_categoria_tienda4.rename("Tienda 4")
], axis=1).fillna(0)

# (opcional) ordena categorías alfabéticamente
df_facturacion_plot = df_facturacion_plot.sort_index()

# Paleta distinta para este gráfico
colores_facturacion = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728',
                       '#9467bd', '#8c564b', '#e377c2', '#7f7f7f',
                       '#bcbd22', '#17becf']

# Graficar (tiendas en el eje X, categorías apiladas)
plt.figure(figsize=(12, 7))
plot_data = df_facturacion_plot.T  # filas: tiendas | columnas: categorías
ax = plot_data.plot(kind="bar", stacked=True, color=colores_facturacion, figsize=(12, 7))

plt.title("Facturación por categoría en cada tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación ($)")
plt.legend(title="Categoría del producto", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=0)
plt.tight_layout()

# Etiquetas $ dentro de cada bloque
for i, tienda_nombre in enumerate(plot_data.index):
    y_offset = 0
    for valor in plot_data.loc[tienda_nombre]:
        if valor > 0:
            ax.text(i, y_offset + valor/2, f"${valor:,.0f}",
                    ha='center', va='center', fontsize=8)
            y_offset += valor

plt.show()
plt.close()
```

# 3. Calificación promedio de la tienda


```python
Promedio_calificacion_tienda1 = tienda["Calificación"].mean()
Promedio_calificacion_tienda2 = tienda2["Calificación"].mean()
Promedio_calificacion_tienda3 = tienda3["Calificación"].mean()
Promedio_calificacion_tienda4 = tienda4["Calificación"].mean()

print('Calificación promedio por tienda \n')
print(f"Calificación promedio tienda 1: {Promedio_calificacion_tienda1:.2f}")
print(f"Calificación promedio tienda 2: {Promedio_calificacion_tienda2:.2f}")
print(f"Calificación promedio tienda 3: {Promedio_calificacion_tienda3:.2f}")
print(f"Calificación promedio tienda 4: {Promedio_calificacion_tienda4:.2f}")


```

```python
# Gráfico de calificaciones promedio por tienda
calificaciones = [Promedio_calificacion_tienda1, Promedio_calificacion_tienda2,
                  Promedio_calificacion_tienda3, Promedio_calificacion_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(6,4))
barras = plt.barh(tiendas, calificaciones, color=['#FF6B6B','#4ECDC4','#FFD93D','#1A535C'])

# Etiquetas en cada barra
for i, v in enumerate(calificaciones):
    plt.text(v + 0.02, i, f"{v:.2f}", va='center', fontsize=9)

plt.title("Calificación promedio por tienda")
plt.xlabel("Calificación promedio")
plt.xlim(0, 5)  # porque las calificaciones están entre 1 y 5
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def productos_extremos(df, nombre_tienda):
    productos = df['Producto'].value_counts()
    mas_vendido = productos.idxmax()
    menos_vendido = productos.idxmin()
    print(f"\n{nombre_tienda.upper()}:\n")
    print(f"🟢 Producto más vendido: {mas_vendido},{productos.max()} vendidos")
    print(f"🔴 Producto menos vendido: {menos_vendido},{productos.min()} vendidos")


productos_extremos(tienda, "Tienda 1")
productos_extremos(tienda2, "Tienda 2")
productos_extremos(tienda3, "Tienda 3")
productos_extremos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
Promedio_envio_tienda1 = tienda["Costo de envío"].mean()
Promedio_envio_tienda2 = tienda2["Costo de envío"].mean()
Promedio_envio_tienda3 = tienda3["Costo de envío"].mean()
Promedio_envio_tienda4 = tienda4["Costo de envío"].mean()

print('Envío promedio por tienda \n')
print(f"Envío promedio tienda 1: ${Promedio_envio_tienda1:,.2f}")
print(f"Envío promedio tienda 2: ${Promedio_envio_tienda2:,.2f}")
print(f"Envío promedio tienda 3: ${Promedio_envio_tienda3:,.2f}")
print(f"Envío promedio tienda 4: ${Promedio_envio_tienda4:,.2f}")
```

```python
# Gráfico de torta de envío promedio por tienda
envios = [Promedio_envio_tienda1, Promedio_envio_tienda2,
          Promedio_envio_tienda3, Promedio_envio_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#FF6B6B','#4ECDC4','#FFD93D','#1A535C']

plt.figure(figsize=(6,6))
plt.pie(envios, labels=tiendas, autopct=lambda p: f"${p*sum(envios)/100:,.2f}\n({p:.1f}%)",
        colors=colores, startangle=90, wedgeprops={'edgecolor':'white'})

plt.title("Costo de envío promedio por tienda")
plt.tight_layout()
plt.show()
```
